```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

# Frases comunes que puede recibir el chatbot
frases_usuario = [
    "me duele la cabeza",
    "mi hijo se golpeó la cabeza y está sangrando",
    "tengo dolor en el pecho",
    "estoy vomitando mucho",
    "tengo fiebre muy alta",
    "necesito una ambulancia urgente",
    "qué especialista atiende problemas cardíacos",
    "tengo dolor en el pie desde hace días",
    "mi madre tiene dificultad para respirar",
    "quiero hablar con un doctor ahora mismo",
    "mi hijo se cayó por las escaleras",
    "siento un dolor agudo en el abdomen",
    "tengo visión borrosa y náuseas",
    "me arde al orinar y tengo fiebre",
    "mi papá está inconsciente",
    "estoy teniendo una crisis de ansiedad",
    "mi hijo tiene convulsiones",
    "mi abuela se cayó y no puede moverse",
    "mi corazón late muy rápido sin razón",
    "tengo dolor al respirar",
]

# Respuestas asociadas
respuestas_chatbot = [
    "Puedes tomar un analgésico, pero si el dolor persiste, consulta con un médico.",
    "Acude de inmediato a urgencias, podría ser una lesión grave.",
    "Esto puede ser una emergencia cardíaca, acude a urgencias inmediatamente.",
    "Es importante que te hidrates y vayas a un hospital si no mejoras.",
    "Consulta con un médico, especialmente si la fiebre es mayor a 39°C.",
    "Enviando ambulancia a tu ubicación. Mantén la calma.",
    "Debes consultar con un cardiólogo.",
    "Podrías requerir un ortopedista. Se recomienda una revisión si el dolor continúa.",
    "Llévala al hospital lo más pronto posible. Puede tratarse de una insuficiencia respiratoria.",
    "Un médico general puede ayudarte o derivarte con un especialista.",
    "Busca atención médica inmediata. Esto puede ser una emergencia.",
    "Podría ser apendicitis o una condición grave. Acude al hospital.",
    "Acude a urgencias para una evaluación neurológica inmediata.",
    "Podrías tener una infección urinaria. Acude con un médico general o urólogo.",
    "Llama a urgencias y realiza maniobras básicas de reanimación si sabes cómo.",
    "Respira profundo, busca un lugar tranquilo y pide ayuda si es necesario.",
    "Acude a urgencias inmediatamente. Las convulsiones pueden poner en riesgo la vida.",
    "Llama a emergencias y evita moverla, podría tener una fractura.",
    "Podría tratarse de una arritmia. Consulta a un cardiólogo lo antes posible.",
    "Busca atención médica urgente, podría ser una infección pulmonar.",
]

# Vectorización
vectorizador = TfidfVectorizer()
X = vectorizador.fit_transform(frases_usuario)

# Función de respuesta usando similitud de coseno
def responder_chatbot(mensaje_usuario):
    mensaje_vectorizado = vectorizador.transform([mensaje_usuario])
    similitudes = cosine_similarity(mensaje_vectorizado, X)
    idx_respuesta = np.argmax(similitudes)
    return respuestas_chatbot[idx_respuesta]

# Prueba
mensaje = "mi mamá está respirando muy mal y se queja del pecho"
respuesta = responder_chatbot(mensaje)
print("Respuesta del chatbot:", respuesta)
```
