```python
from flask import Flask, request, jsonify
from datetime import datetime
import json
import os

app = Flask(__name__)

# Archivo de registro simulado
LOGS_FILE = "llamadas_logs.json"

# Funci√≥n para cargar historial
def cargar_logs():
    if os.path.exists(LOGS_FILE):
        with open(LOGS_FILE, "r") as f:
            return json.load(f)
    return []

# Guardar nueva llamada
def guardar_logs(data):
    with open(LOGS_FILE, "w") as f:
        json.dump(data, f, indent=2)

# Simulaci√≥n de una llamada al hospital
def simular_llamada(hospital, especialidad, nombre_paciente, motivo):
    now = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
    log = {
        "timestamp": now,
        "hospital": hospital,
        "especialidad": especialidad,
        "paciente": nombre_paciente,
        "motivo": motivo,
        "estado": "‚úÖ Llamada simulada exitosa",
        "receptor": "Asesor m√©dico especializado"
    }
    logs = cargar_logs()
    logs.append(log)
    guardar_logs(logs)
    return log

# Endpoint para simular llamada
@app.route("/api/llamada", methods=["POST"])
def api_llamada():
    datos = request.get_json()
    hospital = datos.get("hospital")
    especialidad = datos.get("especialidad")
    nombre = datos.get("nombre")
    motivo = datos.get("motivo")

    if not all([hospital, especialidad, nombre, motivo]):
        return jsonify({"error": "‚ùóFaltan datos necesarios"}), 400

    resultado = simular_llamada(hospital, especialidad, nombre, motivo)
    return jsonify({"mensaje": "üìû Llamada simulada realizada", "detalle": resultado}), 200

# Endpoint para historial de llamadas
@app.route("/api/historial_llamadas", methods=["GET"])
def historial_llamadas():
    logs = cargar_logs()
    return jsonify({"historial": logs}), 200

if __name__ == "__main__":
    app.run(host="0.0.0.0", port=5000)
```
