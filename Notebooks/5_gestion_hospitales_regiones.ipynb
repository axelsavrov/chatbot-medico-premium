```python
# Notebook 5 ‚Äî Emergencias con IA + NLP para interpretar s√≠ntomas

import random
from datetime import datetime
import spacy

# Cargar el modelo en espa√±ol
nlp = spacy.load("es_core_news_sm")

# Diccionario de especialidades con palabras clave
especialidades_nlp = {
    "neurolog√≠a": ["cabeza", "mareo", "desmayo", "convulsi√≥n", "golpe", "traumatismo"],
    "urgencias": ["sangrado", "accidente", "emergencia", "herida", "trauma", "auxilio"],
    "ortopedia": ["fractura", "hueso", "esguince", "pierna", "brazo", "mu√±eca"],
    "cardiolog√≠a": ["coraz√≥n", "infarto", "palpitaciones", "dolor en el pecho", "presi√≥n alta"],
    "pediatr√≠a": ["ni√±o", "beb√©", "infante", "menor", "fiebre", "llora"],
    "medicina interna": ["dolor", "v√≥mito", "fiebre", "diarrea", "malestar", "tos"]
}

# Base de datos de doctores (simplificada)
doctores_disponibles = {
    "neurolog√≠a": ["Dr. Juan Ram√≠rez ‚Äì √Ångeles del Pedregal", "Dra. Ana Torres ‚Äì M√©dica Sur"],
    "urgencias": ["Dr. Luis Casta√±eda ‚Äì Star M√©dica", "Dra. Paula Mendoza ‚Äì Hospital ABC"],
    "ortopedia": ["Dr. Mart√≠n Salazar ‚Äì Hospital Espa√±ol", "Dra. Laura M√©ndez ‚Äì Cl√≠nica Mayo"],
    "cardiolog√≠a": ["Dr. Diego Paredes ‚Äì Cleveland Clinic", "Dra. Sophia Kim ‚Äì Massachusetts General"],
    "pediatr√≠a": ["Dra. Mariana Ruiz ‚Äì H√¥pital Necker", "Dr. Julien Fabre ‚Äì Johns Hopkins"],
    "medicina interna": ["Dr. Eduardo N√∫√±ez ‚Äì Centro M√©dico Nacional", "Dra. Claudia Leal ‚Äì Hospital ABC"]
}

# Funci√≥n de NLP para determinar especialidad
def interpretar_especialidad(mensaje):
    doc = nlp(mensaje.lower())
    tokens = [token.lemma_ for token in doc]

    for especialidad, palabras in especialidades_nlp.items():
        for palabra in palabras:
            if palabra in tokens or palabra in mensaje:
                return especialidad
    return None

# Flujo del chatbot para emergencias
def procesar_emergencia(texto_usuario):
    print("üß† Analizando los s√≠ntomas del paciente...")
    especialidad = interpretar_especialidad(texto_usuario)

    if not especialidad:
        print("‚ùó No se pudo detectar la especialidad. Un asesor humano ser√° asignado.")
        return

    doctor = random.choice(doctores_disponibles[especialidad])
    hora_actual = datetime.now().strftime("%H:%M")

    print(f"‚úÖ Especialidad detectada: *{especialidad.upper()}*")
    print(f"üë®‚Äç‚öïÔ∏è M√©dico asignado: {doctor}")
    print(f"üìû Un asesor te llamar√° en breve. Hora actual: {hora_actual}")
    print("üìã Procederemos a verificar con la aseguradora... ‚úÖ")

# --- PRUEBA DEL SISTEMA ---
mensaje = "Mi hijo se golpe√≥ la cabeza y est√° sangrando mucho"
procesar_emergencia(mensaje)
```
