```python
import spacy
from random import choice
from time import sleep

# Cargar modelos en espa√±ol e ingl√©s
nlp_es = spacy.load("es_core_news_sm")
nlp_en = spacy.load("en_core_web_sm")

# Diccionario de s√≠ntomas a especialidades (b√°sico y expandible)
especialidades = {
    "dolor de cabeza": "neurolog√≠a",
    "fiebre": "medicina interna",
    "pecho": "cardiolog√≠a",
    "coraz√≥n": "cardiolog√≠a",
    "abdomen": "gastroenterolog√≠a",
    "pierna": "ortopedia",
    "pulmones": "neumolog√≠a",
    "depresi√≥n": "psiquiatr√≠a",
    "visi√≥n": "oftalmolog√≠a",
    "respirar": "neumolog√≠a",
    "fractura": "traumatolog√≠a",
    "corte": "urgencias",
    "infecci√≥n": "infectolog√≠a",
    "embarazo": "ginecolog√≠a",
    "piel": "dermatolog√≠a",
    "est√≥mago": "gastroenterolog√≠a",
    "ansiedad": "psiquiatr√≠a",
    "sangrado": "emergencias"
}

# Hospitales premium por pa√≠s
hospitales = {
    "mexico": ["Hospital √Ångeles del Pedregal", "M√©dica Sur", "Hospital Espa√±ol", "Star M√©dica"],
    "usa": ["Mayo Clinic", "Cleveland Clinic", "Mount Sinai"],
    "francia": ["H√¥pital Necker", "Piti√©-Salp√™tri√®re", "Clinique Ambroise Par√©"],
    "japon": ["St. Luke's International Hospital", "Tokyo Medical University Hospital"],
    "colombia": ["Fundaci√≥n Santa Fe", "Cl√≠nica del Country"]
}

# Simulaci√≥n de aseguradora
aseguradora_aprobada = ["AXA", "GNP", "MetLife"]

# Simulaci√≥n de IA para clasificar s√≠ntomas
def detectar_especialidad(texto, idioma="es"):
    nlp = nlp_es if idioma == "es" else nlp_en
    doc = nlp(texto.lower())
    for token in doc:
        for sintoma in especialidades:
            if sintoma in token.text:
                return especialidades[sintoma]
    return "medicina general"

# --- Simulaci√≥n de conversaci√≥n ---

print("ü§ñ Asistente M√©dico Premium - IA de Clasificaci√≥n de Emergencias\n")
idioma = input("Idioma / Language? (es/en): ").strip().lower()
nombre = input("üë§ ¬øCu√°l es tu nombre?: ")
ubicacion = input("üìç ¬øEn qu√© pa√≠s est√°s?: ").strip().lower()
hospital = choice(hospitales.get(ubicacion, ["Hospital General Premium"]))
sintomas = input("üìù Describe qu√© s√≠ntomas tienes o qu√© pas√≥: ")

print("\nüß† Analizando tu situaci√≥n con IA...")
sleep(1.5)
especialista = detectar_especialidad(sintomas, idioma)

print(f"\n‚úÖ {nombre}, basado en tu mensaje te sugerimos un especialista en *{especialista}*.")

# Verificaci√≥n de aseguradora
aseguradora = input("üè• ¬øCu√°l es tu aseguradora?: ").strip()
print("üîç Verificando aseguradora con red hospitalaria...")
sleep(2)

if aseguradora.lower() in [a.lower() for a in aseguradora_aprobada]:
    medico = f"Dr. {choice(['Juan P√©rez', 'Laura G√≥mez', 'Andr√©s R√≠os', 'Elena Mart√≠nez'])}"
    print(f"‚úÖ Aseguradora verificada. El {medico} te atender√° en *{hospital}*.\n")
else:
    print("‚ùå Tu aseguradora no est√° verificada. Por favor paga en recepci√≥n o contacta soporte.")

# Opci√≥n de contacto
llamada = input("üìû ¬øDeseas que un asesor te llame? (s√≠/no): ").strip().lower()
if llamada in ["s√≠", "si", "yes"]:
    print("üì≤ En breve un asesor humano te llamar√°. Gracias por confiar en nosotros.")
else:
    print("üì® Puedes continuar tu atenci√≥n v√≠a WhatsApp o agendar otra cita m√°s adelante.")
```
